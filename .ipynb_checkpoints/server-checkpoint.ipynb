```python
import cv2
import numpy as np
import display
import time
import http.server
from urllib.parse import urlparse
from dataset import *
```

```python
def streamTest():
    ip = "25.21.27.22"
    site = "http://"+ip+":8090/?action=stream"
    cap = cv2.VideoCapture(site)
    while True :
        ret, frame = cap.read()
        if ret:
            display.vidshow(frame)
            time.sleep(0.2)
        else:
            print("error")
            break
    cap.release()
# streamTest()
```

```python
class CatServer:
    def __init__(self, site=None, refCat = "lulu.jpg"):
        self.id = 0
        self.cap = None
        
        ip = "25.21.27.22"
        if site is None:
            self.site = "http://"+ip+":8090/?action=stream"
        else:
            self.site = site
        self.lastRetrive = 0
        
        self.yolo = self.landmark = self.recog = None
        
        self.refCat = refCat
        
        self.thresh = 1.65
        self.unlocked = False
        self.lockUnlockHistory = []
        self.lockUnlockFrameThresh = 3
        self.lockUnlockDuration = 20
    
    def retrive(self, retry=5):
        if self.cap is None or retry<0:
            print("retrive FAILED!")
            return None
        retry -= 1
        ret, frame = self.cap.read()
        if ret:
            self.lastRetrive = time.time()
            return frame
        else:
            print("retrive failed! sleep...", retry)
            time.sleep(0.5)
            self.start()
            return self.retrive(retry)
    
    def start(self):
        print("init camera")
        if not self.cap is None:
            print("capture release")
            self.cap.release()
        self.cap = cv2.VideoCapture(self.site)
        print("load models")
        if self.yolo is None:
            self.yolo = CatYoloDetector(useMostProb=True)
        if self.landmark is None:
            self.landmark = CatLandmarkDetector()
        if self.recog is None:
            self.recog = CatRecognizor()
        print("ref cat")
        img = cv2.imread(self.refCat)
        faces = self.yolo.detect(img)
        if len(faces) < 1:
            print("no face in reference!")
            return
        faces = self.landmark.detect(img, faces)
        ref_face = cv2.medianBlur(img, 1+2*2)
        display.imshow(ref_face)
        faces = self.recog.detect(img, faces)
        self.refFace = faces[0]
        print(self.refFace.vec)
        print("end")
    
    def tryUnlock(self, result):
        tick = time.time()
        self.lockUnlockHistory.append([tick, result])
        i=0
        score = 0
        while i < len(self.lockUnlockHistory):
            item = self.lockUnlockHistory[i]
            if tick - item[0] > self.lockUnlockDuration:
                self.lockUnlockHistory.pop(i)
                i -= 1
            else:
                if item[1]:
                    score += 1
            i+=1
        newlock = score >= self.lockUnlockFrameThresh
        if newlock!=self.unlocked:
            if newlock:
                for _ in range(5):
                    print("UNLOOOOOOOOOOOOOOOOOOOOOOOOOOOOOCKED!")
            else:
                print("lock score..", score)
        self.unlocked = newlock
    
    def serialize(self):
        frame = self.retrive()
        message = str(self.id) + "\n"
        self.id += 1
        if frame is None:
            message += "CameraError\n"
        else:
            faces = self.yolo.detect(frame)
            #cats
            if len(faces) > 0:
                faces = self.landmark.detect(frame, faces)
                faces = self.recog.detect(frame, faces)
                for face in faces:
                    message += "Cat("
                    refFace = self.refFace
                    diff = self.recog.dist(refFace.vec, face.vec)[0]
                    unlocked = diff < self.thresh
                    message += str(diff) + "|"
                    message += str(unlocked) + "|"
                    message += str([face.ax, face.ay, face.aw, face.ah]) + "|"
                    message += str(face.landmarks) + ")\n"
                    face.draw(frame)
                    self.tryUnlock(unlocked)
                display.imshow(frame)
                if self.id % 200 == 0:
                    display.clear()
                print(message)
            else:
                message += "NoCat\n"
            #unlocked
            self.tryUnlock(False)
            if self.unlocked:
                    message += "Unlocked\n"
                else:
                    message += "Locked\n"
        return message

content = CatServer()
content.start()

class MyHandler(http.server.BaseHTTPRequestHandler):
    def do_GET(self):
        parsed_path=urlparse(self.path)
        message=content.serialize() #.join(message_parts)
        content.id += 1
        self.send_response(200) #응답코드
        self.end_headers() #헤더가 본문을 구분
        self.wfile.write(message.encode('utf-8'))
        return

s=http.server.HTTPServer(('',3206),MyHandler)
print("Start serving")
s.serve_forever()
```
